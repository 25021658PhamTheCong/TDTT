```python
lst=list(map(int,input().split()))
n=int(input())
lst_ans = []
for i in range(0, len(lst)):
    for j in range(i + 1, len(lst)) :
        if lst[i] + lst[j] == n :
            if (lst[i], lst[j]) not in lst_ans:
                lst_ans.append((lst[i], lst[j]))
print(lst_ans)
    
```

```python
lst=list(map(int,input().split()))
counts={}
for i in lst:
    counts[i]= counts.get(i, 0) + 1
max_freq=max(counts.values())
candidates=[]
for i in counts:
    if counts[i]==max_freq:
        candidates.append(i)
result=min(candidates)
print(result)
```

```python
#W6A19
def timvitri (lst):
    vitri=0
    for i in range(1, len(lst)):
        if lst[i] > lst[vitri]:
            vitri=i
    return vitri
lst=list(map(int,input().split()))
print(timvitri(lst)+1)
```

```python
#W6A20
def timvitri(lst,k):
    a=-1
    for i in range (0, len(lst)):
        if lst[i]==k:
            a=i+1
            return a
    return a
    
lst=list(map(int,input().split()))
k=int(input())
print(timvitri(lst,k))
```
